```python
dummies = pd.get_dummies(df["Building Type"]).astype(int)
dummies
```

```python
dummies2 = pd.get_dummies(df["Day of Week"]).astype(int)
dummies2
```

```python
merged = pd.concat([df,dummies,dummies2],axis='columns')
merged
```

```python
final = merged.drop(['Building Type','Day of Week'],axis='columns')
final
```

```python
final = final.drop(['Residential','Weekend'],axis='columns')
final
```

```python
X = final.drop(['Energy Consumption'],axis='columns')
X
```

```python
Y = df[["Energy Consumption"]]
Y
```

# Checking for train dataset

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X,Y)
reg
```

```python
reg.score(X,Y)
```
